````python
#1 
user_products = input("add a list of products separated by ,")
try:
    products_list = user_products.strip().split(",")
except:
    print("error in creating products")

def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory

inventory = initialize_inventory(products_list)     
````

````python
#2 Modify the function get_customer_orders so it prompts the user to enter the number of customer orders and gathers 
#the product names using a loop and user input.
#Use comprehension

def get_customer_orders():
    customer_orders = set()
    order_number = input("Enter the number of customer orders: ")
    while not order_number.isdigit():
        order_number = input("Enter the number of customer orders: ")
    int_order_number = int(order_number)
    print(f"int_order_number {int_order_number}")
 
    while len(customer_orders) < int_order_number:       
        customer_orders = {product for product in products_list if product == (input(f"Enter the name of a product that a customer wants to order: {product}")) }       
    return customer_orders

customer_orders = get_customer_orders()

total_available = sum (quantity for item,quantity in inventory.items()  if item in customer_orders)
total_order = len(customer_orders)
if total_available > 0:
    percentage = (total_order / total_available) * 100
else:
    percentage = 0

#Percentage of Unique Products Ordered: 40.0
print("Order Statistics:")
print(f"Total Products Ordered: {total_order}")
print(f"Percentage of Unique Products Ordered: {percentage}")

     
````

````python
#3 Add a new function to calculate the total price of the customer order. 
#For each product in customer_orders, prompt the user to enter the price of that product. 
#Use comprehension to calculate the total price. Note: assume that the user can only have 1 unit of each product.
def calculate_total_price():
    total_price = sum( int(input(f"Enter the Enter the price of: {element}")) for element in customer_orders )  
    return total_price
````

````python
#4  Modify the update_inventory function to remove the product from the inventory 
#if its quantity becomes zero after fulfilling the customer orders. 
#Use comprehension to filter out the products with a quantity of zero from the inventory.

def update_inventory (customer_orders, inventory):         
    for order in customer_orders:
        inventory[order] -= 1
        filtered_inventory= dict(filter(lambda item: item[1] > 0, inventory.items()))
    return filtered_inventory    
    

#Updated Inventory:
#t-shirt: 5
#mug: 4
#hat: 2
           
filtered_inventory= update_inventory(customer_orders, inventory)
for item,value in filtered_inventory.items():
    print(f"remaining Item : {item} : {value}")

````

````python
#5 Print the total price of the customer order.
print (f"Total Price: {calculate_total_price()}")
````
